# Normal distribuition

A normal distribution has 4 characteristics. 


1.  It's symmetrical.
2.  It's unimodal 
3. its asymptotics, which means that as you add more data to your distribution. It stays within a limit of a Bell shaped curve.
4. Mean, median and mode are equal

We know that the region between the average minus 1x the standard deviation and the average plus 1x the standard deviation is where 68% of the data is, while the region between and the average minus 2x the standard deviation and the average plus 2x the standard deviation is where 95% of the data remains.

You can use both histograms an QQ plots and we'll talk about that, further so let's get started.

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

vals = np.random.normal(size=1000)
sns.distplot(vals, hist=False)

plt.show()
```
